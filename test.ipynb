```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

def f(x):
    return x**2

N = 100
x = np.linspace(-10,10,N)


integrals = []
x_range = []
y_range = []
for i in x:
    x_range.append(i)
    y_range.append(f(i))
    integral = integrate.simps(y_range, x_range)
    integrals.append(integral)

plt.plot(x, integrals)
plt.show()
```
